```python
import os
import pandas as pd

os.chdir('/home/satwika/Downloads/archive (1)')

# Load only 100,000 rows from each file instead of full file
oct_df = pd.read_csv('2019-Oct.csv', nrows=100000)
nov_df = pd.read_csv('2019-Nov.csv', nrows=100000)

df = pd.concat([oct_df, nov_df], ignore_index=True)

print("Done! Shape:", df.shape)
print(df.head())
```

```python
# Drop duplicates and missing values
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)

# Convert event_time to datetime
df['event_time'] = pd.to_datetime(df['event_time'])

print("Cleaned shape:", df.shape)
print("\nEvent types:\n", df['event_type'].value_counts())
```

```python
# Count unique users at each funnel stage
visitors  = df[df['event_type'] == 'view']['user_id'].nunique()
leads     = df[df['event_type'] == 'cart']['user_id'].nunique()
customers = df[df['event_type'] == 'purchase']['user_id'].nunique()

print(f"Visitors  : {visitors}")
print(f"Leads     : {leads}")
print(f"Customers : {customers}")
```

```python
visit_to_cart    = (leads / visitors) * 100
cart_to_purchase = (customers / leads) * 100
overall          = (customers / visitors) * 100

print(f"Visit → Cart      : {visit_to_cart:.2f}%")
print(f"Cart → Purchase   : {cart_to_purchase:.2f}%")
print(f"Overall Conversion: {overall:.2f}%")
```

```python
import matplotlib.pyplot as plt

stages = ['Visitors', 'Cart (Leads)', 'Customers']
counts = [visitors, leads, customers]
colors = ['#4C72B0', '#DD8452', '#55A868']

plt.figure(figsize=(8, 5))
plt.bar(stages, counts, color=colors)
plt.title('Marketing Funnel - User Drop-off')
plt.ylabel('Number of Unique Users')
for i, v in enumerate(counts):
    plt.text(i, v + 100, str(v), ha='center', fontweight='bold')
plt.tight_layout()
plt.savefig('funnel_chart.png')
plt.show()
print("Chart saved!")
```

```python
top_cat = df[df['event_type'] == 'purchase']['category_code'].value_counts().head(10)
print(top_cat)

top_cat.plot(kind='bar', figsize=(10, 5), color='steelblue')
plt.title('Top 10 Categories by Purchase')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('top_categories.png')
plt.show()
```

```python
drop1 = visitors - leads
drop2 = leads - customers

print(f"Drop-off Visitors → Cart    : {drop1} users ({(drop1/visitors)*100:.1f}%)")
print(f"Drop-off Cart → Purchase    : {drop2} users ({(drop2/leads)*100:.1f}%)")
```

```python
df.to_csv('cleaned_funnel_data.csv', index=False)
print("Cleaned data saved!")
```

```python
# Monthly comparison
df['month'] = df['event_time'].dt.month_name()

monthly = df.groupby(['month', 'event_type'])['user_id'].nunique().unstack()
print(monthly)

monthly.plot(kind='bar', figsize=(10, 5))
plt.title('Monthly Funnel Comparison (Oct vs Nov)')
plt.ylabel('Unique Users')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('monthly_comparison.png')
plt.show()
```

```python
# Top brands driving purchases
top_brands = df[df['event_type'] == 'purchase']['brand'].value_counts().head(10)
print(top_brands)

top_brands.plot(kind='bar', figsize=(10, 5), color='coral')
plt.title('Top 10 Brands by Purchase')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('top_brands.png')
plt.show()
```
